```python
from summac.model_summac import SummaCZS, SummaCConv
model_conv = SummaCConv(models=["vitc"], bins='percentile', granularity="document", nli_labels="e", device="cpu", start_file="default", agg="mean")
```

```python
model_zs = SummaCZS(granularity="sentence", model_name="vitc", device="cpu")
```

```python

for document, sum in zip(full_text, c0):
  score_conv1 = model_conv.score([document], [sum])
  print("[C0]SummacConv score: %.3f" % (score_conv1["scores"][0]))
```

```python
for document, sum in zip(full_text, cg):
  score_cg = model_conv.score([document], [sum])
  print("[Cg]SummacConv score: %.3f" % (score_cg["scores"][0]))
```

```python
for document, sum in zip(full_text, g0):
  score_g0 = model_conv.score([document], [sum])
  print("[G0]SummacConv score: %.3f" % (score_g0["scores"][0]))

for document, sum in zip(full_text, gp):
  score_gp = model_conv.score([document], [sum])
  print("[Gp]SummacConv score: %.3f" % (score_gp["scores"][0]))

for document, sum in zip(full_text, gg):
  score_gg = model_conv.score([document], [sum])
  print("[Gg]SummacConv score: %.3f" % (score_gg["scores"][0]))
```
