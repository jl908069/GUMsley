### Prepare the data

```python
with open("./data/G4_c0.txt") as f:
    c0 = [line.strip() for line in f]

with open("./data/G4_b0.txt") as f:
    b0 = [line.strip() for line in f]

with open("./data/G4_Pred_b.txt") as f:
    pred_b = [line.strip() for line in f]

with open("./data/G4_Pred_c.txt") as f:
    pred_c = [line.strip() for line in f]

with open("./data/G4_Gold_b.txt") as f:
    gold_b = [line.strip() for line in f]

with open("./data/G4_Gold_c.txt") as f:
    gold_c = [line.strip() for line in f]

with open("./data/GUM9_Gold_Summaries.txt") as f:
    gold = [line.strip() for line in f]

with open("./data/GUM9_brio_summaries.txt") as f:
    brio = [line.strip() for line in f]
```

```python
# Number of documents in each genre
genre_split=[18, 20, 14, 19, 19, 23, 18, 15, 15, 15, 18, 19]
```

### ROUGE scores

```python
r1=[]
r2=[]
l=[]
for i in range(len(gold)):
  summary= [[gold_c[i]]] #modify here
  reference= [[[gold[i]]]]
  rouge = Pythonrouge(summary_file_exist=False,
                    summary=summary, reference=reference,
                    n_gram=2, ROUGE_SU4=False, ROUGE_L=True,
                    recall_only=False, stemming=True, stopwords=True,
                    word_level=True, length_limit=True, length=50,
                    use_cf=True, cf=95, scoring_formula='best',
                    resampling=True, samples=1000, favor=True, p=0.5)
  scores = rouge.calc_score()
  r1.append(scores['ROUGE-1-F'])
  r2.append(scores['ROUGE-2-F'])
  l.append(scores['ROUGE-L-F'])
```

```python
r1_split = [r1[x - y: x] for x, y in zip(
        accumulate(genre_split), genre_split)]

r1_avg=[sum(sub_list) / len(sub_list) for sub_list in r1_split]
print(r1_avg)
#print(len(r1_avg))
```

```python
r2_split = [r2[x - y: x] for x, y in zip(
        accumulate(genre_split), genre_split)]

r2_avg=[sum(sub_list) / len(sub_list) for sub_list in r2_split]
print(r2_avg)
#print(len(r2_avg))
```

```python
l_split = [l[x - y: x] for x, y in zip(
        accumulate(genre_split), genre_split)]

l_avg=[sum(sub_list) / len(sub_list) for sub_list in l_split]
print(l_avg)
#print(len(l_avg))
```

```python
# For better display (dataframe)
sc=[r1_avg,r2_avg,l_avg]
df=pd.DataFrame(sc, index=['rouge1', 'rouge2', 'rougeL']).T
df
```
